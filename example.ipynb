```python
from kse801.abc import load_location

DATA = load_location()
DATA.head()
```

```python
import numpy as np 

latlon_rad = np.radians(
    DATA.loc[:, ['latitude', 'longitude']].to_numpy()
)
timestamps = DATA.loc[:, 'timestamp'].values
```

```python
from poi import PoiCluster

cluster = PoiCluster(
    d_max=100, r_max=250, t_max=60 * 60 * 1000, t_min=5 * 60 * 1000
).fit(
    X=latlon_rad, 
    timestamps=timestamps
)

labels = cluster.predict(X=latlon_rad)
```

```python
DATA_CLUSTERED = DATA.assign(
    cluster=labels
)
DATA_CLUSTERED.head()
```

```python
import altair as alt

alt.data_transformers.disable_max_rows()

alt.Chart(DATA_CLUSTERED).mark_point().encode(
    x=alt.X('latitude:Q', scale=alt.Scale(domain=(36.3, 36.5))),
    y=alt.Y('longitude:Q', scale=alt.Scale(domain=(127.3, 127.5))),
    color=alt.Color('cluster:N')
).interactive()
```
